```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
```

```python
df2.sample(5)
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
X = df2.drop('Churn',axis='columns')
y = df2['Churn']

from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_values = list(range(1, 26))
cv_scores_3 = []
cv_scores_5 = []
cv_scores_7 = []
cv_scores_9 = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores_for_cv_3 = cross_val_score(knn, X, y, cv=3, scoring='accuracy')
    cv_scores_3.append(np.mean(scores_for_cv_3))
    # For 5 fold
    knn = KNeighborsClassifier(n_neighbors=k)
    scores_for_cv_5 = cross_val_score(knn, X, y, cv=5, scoring='accuracy')
    cv_scores_5.append(np.mean(scores_for_cv_5))
    #  For 7 fold
    knn = KNeighborsClassifier(n_neighbors=k)
    scores_for_cv_7 = cross_val_score(knn, X, y, cv=7, scoring='accuracy')
    cv_scores_7.append(np.mean(scores_for_cv_7))
    # For 9 fold
    knn = KNeighborsClassifier(n_neighbors=k)
    scores_for_cv_9 = cross_val_score(knn, X, y, cv=9, scoring='accuracy')
    cv_scores_9.append(np.mean(scores_for_cv_9))

# Find the best k value
best_k_cv_3 = k_values[np.argmax(cv_scores_3)]
best_k_cv_5 = k_values[np.argmax(cv_scores_5)]
best_k_cv_7 = k_values[np.argmax(cv_scores_7)]
best_k_cv_9 = k_values[np.argmax(cv_scores_9)]
best_score_cv_3 = max(cv_scores_3)
best_score_cv_5 = max(cv_scores_5)
best_score_cv_7 = max(cv_scores_7)
best_score_cv_9 = max(cv_scores_9)

print("3 fold cross-validation")
print("Best k:", best_k_cv_3)
print("Cross-validation accuracy with best k:", best_score_cv_3)

print("5 fold cross-validation")
print("Best k:", best_k_cv_5)
print("Cross-validation accuracy with best k:", best_score_cv_5)

print("7 fold cross-validation")
print("Best k:", best_k_cv_7)
print("Cross-validation accuracy with best k:", best_score_cv_7)

print("9 fold cross-validation")
print("Best k:", best_k_cv_9)
print("Cross-validation accuracy with best k:", best_score_cv_9)
```
